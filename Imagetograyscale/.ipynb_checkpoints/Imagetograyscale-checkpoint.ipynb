```python
from PIL import Image
```

```python
in_img = 'input.jpg'
out_img = 'output.jpg'
```

```python
#open Image
with Image.open(in_img) as img:
    #convert to grayscale
    grayscale_img = img.convert('L')
    #save grayscale_img
    grayscale_img.save(out_img)
```
